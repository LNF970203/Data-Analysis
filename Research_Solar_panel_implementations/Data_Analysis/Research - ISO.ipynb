```python
#defining functions
def bar_plot(column_name,x_label_name,title, data = data):
    print(data.shape)
    ax = sns.countplot(x = data[column_name], )
    ax.set_xlabel(f"{x_label_name}")
    ax.set_ylabel("Frequency")
    ax.set_title(f"{title}")
    #getting percentages
    for p in ax.patches:
        ax.annotate('{:.2f}%'.format(p.get_height()/len(data[column_name])*100), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

```python
def pie_chart(column_name, title, legend_title ,explode, data = data):
    print(data.shape)
    assert type(explode) == list, "Explde must be a list"
    df = data.groupby(column_name)[column_name].count()
    df = pd.DataFrame(df)
    df.rename(columns = {f'{column_name}':"Count"}, inplace=True)
    df.reset_index(inplace=True)
    
    plt.pie(x = df["Count"], 
            labels=df[column_name], 
            explode = explode, 
            autopct='%1.2f%%', 
            shadow=True)
    plt.legend(title = f"{legend_title}")
    plt.title(f"{title}")
    plt.show()
    
```
